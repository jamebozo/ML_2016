## Q1
輸入hw0_data.dat，請將指定column由小排到大並印出來。
column的index從0開始
輸出一行，以逗點分隔。ex: 0,0,1,2,3,4,5



```python
import numpy as np
from numpy import loadtxt,savetxt,zeros

def column_sort(column,file):
    lst = []
    data = loadtxt(file)
    for row_lst in data:
        lst.append(row_lst[column])
    lst = sorted(lst,key=float)
    return lst
print (column_sort(3,"hw0_data.dat"))
```
